```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define numerical features
numerical_features = [
    'popularity', 'duration_ms', 'danceability', 'energy', 'key', 'loudness',
    'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature'
]

# Define categorical features
categorical_features = ['explicit', 'mode', 'track_genre', 'popularity_category']

# Set up the matplotlib figure
fig, axes = plt.subplots(len(numerical_features), 1, figsize=(10, 35))

# Draw the numerical feature distributions
for i, feature in enumerate(numerical_features):
    sns.histplot(df[feature], bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {feature.capitalize()}')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Set up the matplotlib figure for categorical features
fig, axes = plt.subplots(len(categorical_features), 1, figsize=(10, 20))

# Draw the categorical feature distributions
for i, feature in enumerate(categorical_features):
    sns.countplot(y=df[feature], ax=axes[i], order = df[feature].value_counts().index)
    axes[i].set_title(f'Distribution of {feature.capitalize()}')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Compute the correlation matrix for numerical features
correlation_matrix = df[numerical_features].corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 9))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)

# Show the plot
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# Set up the matplotlib figure for box plots
fig, axes = plt.subplots(len(numerical_features), 1, figsize=(10, 35))

# Draw the box plots for numerical features against popularity_category
for i, feature in enumerate(numerical_features):
    sns.boxplot(x='popularity_category', y=feature, data=df, ax=axes[i], showfliers=False)
    axes[i].set_title(f'{feature.capitalize()} by Popularity Category')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Identify the top 10 genres based on their frequencies
top_genres = df['track_genre'].value_counts().head(10)

# Set up the matplotlib figure
plt.figure(figsize=(10, 6))

# Draw the bar plot for top 10 genres
sns.barplot(y=top_genres.index, x=top_genres.values, palette='viridis')
plt.title('Top 10 Genres')
plt.xlabel('Count')
plt.ylabel('Genre')

# Show the plot
plt.show()

# Return the top 10 genres
top_genres
```

```python
# Filter the dataset to include only rows where 'popularity_category' is 'Chart-Toppers'
chart_toppers_dataset = df[df['popularity_category'] == 'charttoppers']

# Aggregate the data to get the count of chart-topping songs for each artist
artists_chart_toppers_count = chart_toppers_dataset['artists'].value_counts().reset_index()
artists_chart_toppers_count.columns = ['artists', 'chart_toppers_count']

# Select the top 10 artists with the most chart-topping songs
top_10_artists_chart_toppers = artists_chart_toppers_count.head(10)
top_10_artists_chart_toppers
```

```python
# Calculate the average danceability for each genre
average_danceability_per_genre = df.groupby('track_genre')['danceability'].mean().reset_index()

# Identify the top 10 most frequent genres in the dataset
top_10_frequent_genres = df['track_genre'].value_counts().head(10).index.tolist()

# Filter the average danceability data to include only the top 10 most frequent genres
top_genres_danceability = average_danceability_per_genre[average_danceability_per_genre['track_genre'].isin(top_10_frequent_genres)]

# Sort the data by average danceability in descending order
top_genres_danceability = top_genres_danceability.sort_values(by='danceability', ascending=False)

# Set up the matplotlib figure
plt.figure(figsize=(10, 6))

# Draw the Bar Plot
sns.barplot(x='danceability', y='track_genre', data=top_genres_danceability, palette='viridis')
plt.title('Average Danceability for Top 10 Most Frequent Genres')
plt.xlabel('Average Danceability')
plt.ylabel('Genre')

# Show the plot
plt.show()
```

## What makes a popular song ?

```python
# Grouping by 'popularity_category' and getting the mean of each feature per category
category_mean = df.groupby('popularity_category').mean()

# Sorting the resulting DataFrame by 'popularity' in descending order for better readability
category_mean.sort_values(by='popularity', ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
corr = df.corr()

# Extract correlations with popularity
popularity_corr = corr["popularity"].sort_values(ascending=False)

# Plot correlation matrix focusing on popularity
plt.figure(figsize=(10, 8))
sns.heatmap(corr[['popularity']], annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix (Focus on Popularity)')
plt.show()

popularity_corr
```

```python
import scipy.stats as stats

# Initialising a dictionary to hold p-values for each feature
p_values = {}

# List of features to perform ANOVA on
features = ['explicit', 'duration_ms', 'danceability', 'loudness', 'instrumentalness', 'valence', 'speechiness']

# Performing ANOVA for each feature
for feature in features:
    # Extracting the groups
    group1 = df[df['popularity_category'] == 'charttoppers'][feature]
    group2 = df[df['popularity_category'] == 'mainstreamhits'][feature]
    group3 = df[df['popularity_category'] == 'upandcoming'][feature]
    group4 = df[df['popularity_category'] == 'emerging'][feature]

    # Performing ANOVA and storing the p-value
    f_stat, p_value = stats.f_oneway(group1, group2, group3, group4)
    p_values[feature] = p_value

p_values
```
