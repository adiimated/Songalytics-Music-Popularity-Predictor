```python
# drop unnnamed column - irrelevant
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# dulicate rows
df.duplicated().sum()
```

```python
# data type of the columns
df.dtype
```

```python
# drop null values
df.dropna(subset=['artists', 'album_name', 'track_name'], inplace=True)
```

```python
df.head()
```

```python
# drop duplicates
df.drop_duplicates(inplace=True)
```

```python
df['artists'] = df['artists'].str.strip()

df['album_name'] = df['album_name'].str.strip()

df['track_name'] = df['track_name'].str.strip()
```

```python
# convert boolean data type to int - true or false to 1/0
df['explicit'] = df['explicit'].astype(int)
```

```python
df.shape[0]
```

```python
# drop irrelevant column
df.drop(columns=['track_id'], inplace=True)
```

```python
df.head()
```

```python
# remove outliers
# Calculate the IQR for the 'duration_ms' column
Q1_duration = df['duration_ms'].quantile(0.25)
Q3_duration = df['duration_ms'].quantile(0.75)
IQR_duration = Q3_duration - Q1_duration

# Define bounds for the outliers
lower_bound_duration = Q1_duration - 1.5 * IQR_duration
upper_bound_duration = Q3_duration + 1.5 * IQR_duration

# Filter the dataset to remove outliers in 'duration_ms' column
df_no_outliers = df[(df['duration_ms'] >= lower_bound_duration) & (df['duration_ms'] <= upper_bound_duration)]

# Compare the number of rows before and after removing outliers
before = df.shape[0]
after = df_no_outliers.shape[0]

before, after
```

```python
# added new column - popularity category based on popularity score
import numpy as np

# Define the popularity categories based on the provided ranges
conditions = [
    (df_no_outliers['popularity'] >= 0) & (df_no_outliers['popularity'] <= 25),
    (df_no_outliers['popularity'] >= 26) & (df_no_outliers['popularity'] <= 50),
    (df_no_outliers['popularity'] >= 51) & (df_no_outliers['popularity'] <= 75),
    (df_no_outliers['popularity'] >= 76) & (df_no_outliers['popularity'] <= 100)
]

# Define the corresponding category names
categories = ['emerging', 'upandcoming', 'mainstreamhits', 'charttoppers']

# Create a new column 'popularity_category' in the dataset
df_no_outliers['popularity_category'] = np.select(conditions, categories, default='Unknown')

# Display the first few rows of the modified dataset
df_no_outliers.head()
```

```python
# Define the path to save the cleaned and modified dataset
cleaned_file_path = '/content/drive/MyDrive/Colab Notebooks/spotify/dataset/cleaned_data.csv'

# Save the modified dataset to the specified path
df_no_outliers.to_csv(cleaned_file_path, index=False)
```
